```python
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt

CSV_PATH = "../data/raw/train.csv"
IMG_DIR = "../data/raw/faces-spring-2020/faces-spring-2020"
```

```python
df = pd.read_csv(CSV_PATH)
print(df.head())
```

```python
def show_image(img, label, idx):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(4,4))
    plt.imshow(img_rgb)
    plt.title(f"Index: {idx} | Label: {label}")
    plt.axis("off")
    plt.show()
```

```python
i = 0
```

```python
while i < len(df):
    row = df['id'][i]
    
    img_name = f"face-{row}.png"
    img_path = os.path.join(IMG_DIR, img_name)

    img = cv2.imread(img_path)

    if img is None:
        print(f"Skipping missing: {img_name}")
        i += 1
        continue

    label = df["glasses"][i]
    show_image(img, label, i)
    action = input("Enter: [Enter=correct, f=flip, q=quit] ")

    if action == "f":
        df.at[i, "glasses"] = 1 - label
        print("Flipped label")

    elif action == "q":
        df.to_csv("../data/processed/train.csv", index=False)
        print("Saved cleaned dataset ✅ till index: ", i)
        break

    i += 1
```

After this cell is to manually label the test set in a similar fashion and then concatenate it with the processed training set

```python
df = pd.read_csv("../data/raw/test.csv")
display(df.head())
labels = pd.Series([-1] * len(df))
df["glasses"] = labels
display(df.head())
```

```python
i = 395
```

```python
while i < len(df):
    row = df['id'][i]
    
    img_name = f"face-{row}.png"
    img_path = os.path.join(IMG_DIR, img_name)

    img = cv2.imread(img_path)

    if img is None:
        print(f"Skipping missing: {img_name}")
        i += 1
        continue
    show_image(img, label=df['glasses'][i], idx=i)
    action = input("Enter: [Enter=correct, 0=0, 1=1, q=quit] ")

    if action in {"0", "1"}:
        df.at[i, "glasses"] = int(action)
        print("Updated label")

    elif action == "q":
        print("Saved cleaned dataset ✅ till index: ", i)
        break

    i += 1
```

```python
main_df = pd.read_csv("../data/processed/train.csv")
main_df = pd.concat([main_df, df], ignore_index=True)
main_df.to_csv("../data/processed/train.csv", index=False)
```

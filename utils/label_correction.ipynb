```python
import pandas as pd

#load txt file, change path as required
with open("/kaggle/input/datasets/anvik029/flipper/flipped_1501-3000.txt", "r") as f:
    ids_to_flip = set(int(line.strip()) for line in f if line.strip())

df = pd.read_csv("/kaggle/input/datasets/jeffheaton/glasses-or-no-glasses/train.csv")

mask = df["id"].isin(ids_to_flip)
df.loc[mask, "glasses"] = 1 - df.loc[mask, "glasses"]

#save in a separate writeable path
df.to_csv("/kaggle/working/train.csv", index=False)

print(f"Done. Flipped 'glasses' for {mask.sum()} rows.")
```
